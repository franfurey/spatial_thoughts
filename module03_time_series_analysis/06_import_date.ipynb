```python
### import necessary libraries
import ee
import geemap
import sklearn
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json

ee.Authenticate()
ee.Initialize(project="ee-franciscofurey")
```

```python
taluks = ee.FeatureCollection("users/ujavalgandhi/gee-water-resources/kgis_taluks")

Map = geemap.Map()

Map.addLayer(taluks, {}, "Argentina")

def addArea(f):
    area = f.getNumber('SHAPE_STAr')
    area_sqkm = area.divide(1e6)
    return f.set('area_sqkm', area_sqkm)

taluk_with_area = taluks.map(addArea)


# Fetch the information of the first feature
info = taluk_with_area.first().getInfo()

# Pretty print the information
print(json.dumps(info, indent=2))
```
